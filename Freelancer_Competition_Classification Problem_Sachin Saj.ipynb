```python
import numpy as np
import pandas as pd

from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression

from sklearn import metrics
import time as t
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('training_data.csv')
print('Shape of the dataset -> ',data.shape)
print('Length of the dataset -> ',len(data))
print(data.columns)
data.head(10)
```

```python
X = data.drop(['Target'], axis=1)
Y = data['Target']
```

```python
print(X.shape)
print(Y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
sc = MinMaxScaler()
x_train = sc.fit_transform(x_train)
```

```python
#print(x_train)
#print(y_train)
```

```python
svc = SVC(kernel='sigmoid', gamma=1.0)
knc = KNeighborsClassifier(n_neighbors=49)
mnb = MultinomialNB(alpha=0.2)
dtc = DecisionTreeClassifier(min_samples_split=7, random_state=111)
lrc = LogisticRegression(solver='liblinear', penalty='l1')
rfc = RandomForestClassifier(n_estimators=31, random_state=111)
abc = AdaBoostClassifier(n_estimators=62, random_state=111)
bc = BaggingClassifier(n_estimators=9, random_state=111)
etc = ExtraTreesClassifier(n_estimators=9, random_state=111)

logloss = []
score = []
train_time = []
for i,j in enumerate([svc,knc,mnb,dtc,lrc,rfc,abc,bc,etc]):
    j.fit(x_train,y_train)
    pred = j.predict(x_test)
    logloss.append([metrics.log_loss(y_test,pred)])
    score.append([metrics.accuracy_score(y_test,pred)])
```

```python
ML_Classifier = ['SVM','KNeighborClassifier', 'MultinomialNB',
        'DecisionTreeClassifier','LogisticRegression','RandomForestClassifier',
        'Adaboost','BaggingClassifier','ExtraTreesClassifier']
ML_Classifier = pd.DataFrame(ML_Classifier)
logloss = pd.DataFrame(logloss)
score = pd.DataFrame(score)
ML_Classifier['Log_Loss'] = logloss
ML_Classifier['Score'] = score
ML_Classifier.columns = ['ML_Classifier','Log_Loss','Score']
```

```python
ML_Classifier.head(10)
```

### Conclusion:
- From this table, it is clear that Adaboost ML classifier performed the best with logloss of 0.15
