### Data Cleaning

```python
data.isnull().sum()
```

```python
data_N = data.drop(['IndexSA','AveragePriceSA'], axis=1)
data_N.shape
# Shape of the data after removing two columns
```

```python
Data_C = data_N.dropna(axis=0, how='any',inplace=False)
Data_C.head(5)
# Removing the rows where there is null values
```

```python
Data_C.shape
# Shape of the dataset after removing the null values
```

```python
Data_C.isnull().sum()
# The dataset is clean
```
